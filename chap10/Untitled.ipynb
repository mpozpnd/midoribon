<h3> 階層ベイズモデルのStanによるMCMCサンプリング

```python
%matplotlib inline
import numpy as np
import pystan
from matplotlib import pyplot as plt

X = np.loadtxt("./data10.csv", dtype=np.int)

N = X.shape[0]

stan_data = {'N': N, 'Y': X}

fit = pystan.stan(file='model10.stan', data=stan_data, iter=10000, chains=1)


fit.plot()
plt.show()
```
